```python
import pandas as pd
import numpy as np
import ast
```

```python
movies=pd.read_csv("5000movies.csv")
credits=pd.read_csv("5000credits.csv")
```

<h3>Data Processing</h3>

```python
movies.head(2)
```

```python
credits.head(2)
```

```python
movies=movies.merge(credits,on="title")
```

```python
movies.head(2)
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(2)
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[1].genres
```

```python
movies.iloc[1].keywords
```

```python
movies.iloc[1].cast
```

```python
def standardize(item):
    l=[]
    for i in ast.literal_eval(item):
        l.append(i['name'])
    return l
        
```

```python
movies['genres']=movies["genres"].apply(standardize)
```

```python
movies['keywords']=movies['keywords'].apply(standardize)
```

```python
def standardize2(item):
    l=[]
    count=0
    for i in ast.literal_eval(item):
        if count!=3:
            l.append(i['name'])
            count+=1
        else:
            break    
    return l
```

```python
movies["cast"]=movies["cast"].apply(standardize2)
```

```python
def standardize3(item):
    l=[]
    for i in ast.literal_eval(item):
        if i["job"]=="Director":
            l.append(i['name'])
        else:
            continue
    return l
```

```python
movies["crew"]=movies["crew"].apply(standardize3)
```

```python
movies["overview"]=movies["overview"].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
movies["cast"]=movies["cast"].apply(lambda x:[i.replace(" ","") for i in x])
movies["crew"]=movies["crew"].apply(lambda x:[i.replace(" ","") for i in x] )
movies["genres"]=movies["genres"].apply(lambda x:[i.replace(" ","") for i in x])
movies["keywords"]=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies["tags"]=movies["overview"]+movies["genres"]+movies["cast"]+movies["crew"]+movies["keywords"]
```

```python
new_movies=movies[["movie_id","title","tags"]]
```

```python
new_movies["tags"]=new_movies['tags'].apply(lambda x:" ".join(x))
```

```python
new_movies.head(1)
```

```python
new_movies["tags"]=new_movies["tags"].apply(lambda x:x.lower())
```

<h3>Vectorisation</h3>

<h4>TF-IDF</h4>
<p>TF-IDF or Term Frequency–Inverse Document Frequency, is a numerical statistic that’s intended to reflect how important a word is to a document.</p>

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
new_movies["tags"][4]
```

```python
tfidf=TfidfVectorizer()
```

```python
transformed = tfidf.fit_transform(new_movies['tags'])
```

```python
similarity=cosine_similarity(transformed)
```

```python
similarity
```

```python
list(enumerate(similarity[0]))
```

```python
def get_recommendation(movie):
    movie_index=new_movies[new_movies["title"]==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
        
```

```python
get_recommendation("The Dark Knight Rises")
```
